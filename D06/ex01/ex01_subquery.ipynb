<h2>Подсчитайте, сколько строк удовлетворяют следующим условиям, используя один запрос с любым количеством подзапросов.</h2>

Посчитайте строки в pageviews таблице, но только те из них, checker в которых:


- status = 'ready', мы не хотим анализировать журналы, которые проверяют статус.

- numTrials = 1, мы хотим проанализировать только первые фиксации, поскольку они говорят нам, когда студент начал работать в лаборатории.
«labnames» должны быть из списка: laba04, laba04s, laba05, laba06, laba06s, и project1. Только они были активны во время эксперимента.


- Сохраните результаты проверки в фрейме данных в столбце «cnt».

```python
query = """
SELECT COUNT(*) 
FROM pageviews 
WHERE uid IN (
    SELECT uid
    FROM checker
    WHERE status = 'ready'
      AND numTrials = 1
      AND labname IN ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1')
)
"""
```

```python
result = pd.read_sql(query, conn)
```

```python
print(result)
```
