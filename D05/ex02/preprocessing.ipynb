```python
import pandas as pd
```

Загрузите и прочитайте CSV-файл, создав IDиндексный столбец.

```python
df = pd.read_csv('auto.csv', sep = ',', index_col = 'ID')
```

Подсчитайте количество наблюдений, используя метод count().

```python
df.count()
```

Удалите дубликаты, принимая во внимание только следующие столбцы: CarNumber, Make_n_Model, и Fines.
- Из двух равных наблюдений выберите последнее.
- Проверьте количество наблюдений еще раз.

```python
df = df.drop_duplicates(subset = ['CarNumber', 'Make_n_model', 'Fines'], keep = 'last')
```

```python
df.count()
```

Работа с пропущенными значениями.
- Проверьте, сколько значений отсутствует в каждом столбце.
- Удалите все столбцы с более чем 500 пропущенными значениями, используя аргумент thresh. Проверьте количество пропущенных значений в каждом столбце.
- Замените все пропущенные значения в Refund столбце предыдущим значением в этом столбце для данной ячейки. Используйте метод аргументов и проверьте количество пропущенных значений в каждом столбце.
- Замените все пропущенные значения в Fines столбце средним значением этого столбца (исключая значения NA/NULL при вычислении среднего). 
- Проверьте количество пропущенных значений в каждом столбце.

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 1, thresh = len(df) - 500)

df.isna().sum()
```

```python
df['Refund'] = df['Refund'].fillna(method = 'ffill')

df.isna().sum()
```

```python
df['Fines'] = df['Fines'].fillna(df['Fines'].mean())

df.isna().sum()
```

Разделить и проанализировать марку и модель.
- Используйте метод apply для разделения и извлечения значений в новые столбцы, Makeа также Model.
- Удалить столбец Make_n_Model.
- Сохраните фрейм данных в auto.json файле JSON в следующем формате:
[{"CarNumber":"Y163O8161RUS","Refund":2.0,"Fines":3200.0,"Make":"Ford",
"Model":"Focus"},
{"CarNumber":"E432XX77RUS","Refund":1.0,"Fines":6500.0,"Make":"Toyota",
"Model":"Camry"}]

```python
df[['Make', 'Model']] = df['Make_n_model'].apply(lambda x: pd.Series(str(x).split(' ')))
```

```python
df = df.drop(columns='Make_n_model')
```

```python
df.to_json("auto.json", orient="records", force_ascii=False, indent=4)
```
