Итерации: во всех следующих подзадачах необходимо выполнить вычисления fines/refund*year для каждой строки. Создайте новый столбец с вычисленными данными. Измерьте время с помощью магической команды %%timeit в ячейке.
- Напишите функцию, которая циклически проходит по фрейму данных for i in range(0, len(df)), используя iloc, и , append()к списку. Присвойте результат функции новому столбцу фрейма данных.
- Сделайте это с помощью iterrows().
- Сделайте это с помощью apply() лямбда-функции.
- Сделайте это, используя Series объекты из фрейма данных.
- Сделайте то же, что и в предыдущей подзадаче, но используйте метод .values.

```python
%%timeit

result = []
for i in range(0, len(df)):
    calc = df['Fines'].iloc[i] / df['Refund'].iloc[i] * df['Year'].iloc[i]
    result.append(calc)

df['calc_for'] = result
```

```python
%%timeit

results = []
for idx, row in df.iterrows():
    results.append(row["Fines"] / row["Refund"] * row["Year"])
df["calc_iterrows"] = results
```

```python
%%timeit

df['calc_apply'] = df.apply(lambda x: x['Fines'] / x['Refund'] * x['Year'], axis = 1)
```

```python
%%timeit

df['calc_series'] = df['Fines'] / df['Refund'] * df['Year']
```

```python
%%timeit
df["calc_values"] = df["Fines"].values / df["Refund"].values * df["Year"].values
```

Понижающее приведение:
- Запустите df.info(memory_usage='deep') и обратите внимание на Dtype и использование памяти.
- Преобразуйте copy() свой исходный фрейм данных в другой фрейм данных, optimized_df.
- Понижающее приведение от float64 к float32 для всех столбцов.
- Понижение от int64 до наименьшего возможного числового типа D.
- Запустите info(memory_usage='deep') новый фрейм данных. Обратите внимание на Dtype и использование памяти.

```python
df.info(memory_usage='deep')
```

```python
optimized_df = df.copy()
```

```python
for col in optimized_df.select_dtypes(include="float64"):
    optimized_df[col] = optimized_df[col].astype("float32")
```

```python
for col in optimized_df.select_dtypes(include="int64"):
    optimized_df[col] = optimized_df[col].astype("int16")
```

```python
optimized_df.info(memory_usage='deep')
```

Категории:
- Измените object тип столбцов на category.
- На этот раз проверьте использование памяти. Вероятно, оно уменьшится в 2–3 раза по сравнению с исходным фреймом данных.

```python
for col in optimized_df.select_dtypes('object'):
    optimized_df[col] = optimized_df[col].astype('category')
```

```python
optimized_df.info(memory_usage='deep')
```
